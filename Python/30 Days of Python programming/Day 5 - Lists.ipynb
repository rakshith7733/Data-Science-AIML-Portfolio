```Accessing List Items using Indexing```

- We can access item in a list using inxeding. A index starts from 0.

````python
#access "banglore" using index position
print(lst_1[3])
````

````python
#last index
last_index = len(lst_1) - 1
print(last_index)
````

```Accessing List items using Negative Indexing```

````python
#Access last item
print(lst_1[-1])
````

````python
#access age through negative indexing
print(lst_1[-3])
````

```Unpacking list items```

````python
item = ['item1','item2','item3', 'item4', 'item5']

#*rest is used to assigin remaining items to rest variable.
f_item,sec_item,th_item,*rest = item
print(f_item)
print(sec_item)
print(th_item)
print(rest)
````

```Slicing Items from a list```

- We can specify a range of positive indexes by the start and end and even the step to return value.

- Indexing format:
           [start: stop: step]

````python
fruits = ["Apple","Banana","Grapes",'Orange']

#Positive Indexing
#range 1:3
print(fruits[1:3])

#skip first and pull all items
print(fruits[1:])

#slice the alternate items
print(fruits[::2])
````

````python
#Negative Indexing

#it doesn't include last and goes down
print(fruits[-3:-1])

#-3 to all the way up
print(fruits[-3:])

#Reverse a list using negative index
print(lst_1[::-1])
````

```Modifying the lists```
- Lists are mutable so we can modify the ordered collections of items.

````python
fruits = ["Apple","Banana","Grapes",'Orange']

#Adding the element in index 02
fruits[2] = "Avacado"
print(fruits)
````

````python
#Checking if the element is present in the list
print('Apple' in fruits)
````

```Adding elements in the list using Append() method.```

````python

fruits.append("Strawberry")
print(fruits)
````

```Inserting Items into a list```
- We can insert a specified item on specified element in the list, the other item are shifted to the right side. insert() method.

```python
#Syntax
lst.insert(index,item)
```

````python
fruits.insert(1,'Guava')
print(fruits)
````

```Removing items from a list```
- The remove method remove the specified item from a list.

```python
syntax:
lst.remove(item)
```

````python
#remove item strawberry from the list.
fruits.remove('Strawberry')
print(fruits)
````

```Removing items using pop```
- The pop() method removes the specified index, or the last item if index is not specified.

```py
Syntax:
lst.pop(Index)
```

````python
#Remove "orange" using pop
fruits.pop(4) #Orange
print(fruits)
````

````python
#remove last elemet "Avacado" without giving index position.
fruits.pop()  #Avacado
print(fruits)
````

```Removing item using DEL```
- The del keyword removes the specified index and it can also be used to delete items within index range.
- It can also delete the entire list.

```python
del lst[index] #Deletes a single item

del lst #deletes entire list
```

````python
#delete the first index item
del fruits[1] #Guava
print(fruits)
````

````python
#del fruits #deletes the entire list
````

``Clearning list items```

- The clear() method empties the list

```py
Syntax:
lst.clear()
```

````python
fruits.clear()
print(fruits) #empty list will be produced
````

```Copying a list```
- This will copy the list2 to list1 where list1 = list2. This can be done through copy() method.

```py
Synatx:
lst1 = ['item1','item2']
lst2 = lst1.copy()
```

````python
#Copying lst1 to fruits
fruits_2 = ["Berries",'Watermelon']
fruits = fruits_2.copy()
print(fruits)
````

```Joining Lists```
- Lists can be joined using Join or concatnation (+).

```py
Synatx:
list = list1 + list2
```

````python
vegetables = ['Tomato', 'Potato', 'Cabbage', 'Onion', 'Carrot']

#Concatenate
fruits_and_vegetables = fruits + vegetables
print(fruits_and_vegetables)
````

- Joining using extend() method, the extend allows the list to append it to the end of the list.

```py
Syntax:
list1 = ['item1','item2']
list2 = ['item3','item4']
list1.extend(list2) #['item1','item2','item3','item4']
```

````python
#extending new vegetables to fruits_and_vegetables
citrus_fruit = ['lemon',]
fruits_and_vegetables.extend(citrus_fruit)
print(fruits_and_vegetables)
````

```Counting Items in a list```
- The count() method returns the number of times an item appears in a list.

```py
Synatx:
lst.count(item)
```

````python
ages = [22, 19, 24, 25, 26, 24, 25, 24]
print(ages.count(19)) #1 time
````

```Finding Index of an item```
- The index() method returns the index of an item in the list.

```PY
Syntax:
lst.index(item,start,stop) #{if not found it will result valueerror}
```

````python
print(ages.index(24)) #First Occurance - 2
````

```Reversing a list```

- The reverse() method reverses the order of a list.

```py
Syntax:
lst.reverse()
```

````python
ages.reverse()
print(ages)
````

```Sorting List Items```
- To sort lists items we can use in-built sort() or sorted() method to reorder the list.

sort() - modifies the original list.

```py
Syntax:
lst.sort() #Ascending order
lst.sort(reverse=True) #Descrending order
```

````python
ages.sort()
print(ages) #Ascending order
````

````python
ages.sort(reverse=True)
print(ages) #Descending order
````

- Sorted() returns the ordered list without modifying the original list

````python
fruits_and_vegetables.sort()
print(fruits_and_vegetables)
````
