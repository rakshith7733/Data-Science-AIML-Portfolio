```py
### String Formatting:
- The '%' operator is used to format a set of variables enclosed in a tuple together with a format string.
%s - String(any object with string)
%d - integers
%f - Floating point numbers
%.number of digits f - floating point numbers with fixed precision.

```

````python
# string and numbers
r = 10
pi = 3.14
area = pi * r * r
#Format 1
formatted_string = 'The Area of circle is %d is %.2f.' %(r,area)
print(formatted_string)

python_libraries = ['django','flask','Numpy']
#format 2
formatted_string_1 = 'The widely used python libraries are %s' %(python_libraries)
print(formatted_string_1)

#format 3
age = 26
formatted_string_2 = 'Hello. My name is {} {} and my age is {}'.format(first_name,last_name,age)
print(formatted_string_2)

#format 4
print(f"Are you {full_name}?, I got your application under {age} category.")
````

````python
print("The area of circle with radius {} is {:.2f}.".format(r,area))
````

````python
#Accessing the string charcters & Slicing
print(name[0]) #'p'
print(name[2:]) #thon
print(name[::-1]) #nohtyp - reverse a string.
print(name[-1]) #n
print(name[0:3]) #starting from 0 and not including 3 element.

#[Start:end:step]
print(name[0:4:2]) #pt
````
