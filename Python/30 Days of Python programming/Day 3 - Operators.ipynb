```py
Boolean - A boolean data type represents one or two values to be True or False.

Ex: print(True)
    print(False)

Assignment Operator - These are used to assign a value to the variable, we use this operators to store a value.

Ex: =, +=,-=,*=

Arthemetic Operator:
- These are used to perform action on two variables

Addition, subtraction, multiplication, division, modulus, floor division, exponential

```

````python
#Declaring values and organizing them together
num_one = 3
num_two = 4

#Arthemetic operations
total = num_one + num_two
diff = num_two - num_one
prod = num_one * num_two
div = num_two / num_one
mod = num_two % num_one

print('Total: ',total)
print('Difference: ',diff)
print('Multiplication: ',prod)
print('Division: ',div)
print("Modulus: ",mod)
````

```py
#Comparison Operators
- we use comparision operator to compare two values, to check if values is greater or lesser.

Ex: == (Equal to)
    >= (Greater than equal to)
    <= (less than equal to)
    != (not equal to)
```

````python
#Comparision operator example
print(3>2) #True - Greater than check
print(2<4) #True - less than check
print(4>=4) #True = greater than equal to
print(4<=4) #True - less than equal to
print(len('good') == len('bad')) #False - equal to
print(len('milk') != len('meat')) #False - Not equal to

print(True == True)
print(False == True)
````

```py
IS - return True if both the variables having same object (x is y)
IS NOT - return True if both variables doesn't have same objects (x is not y).
IN - return True if the queried list contains a certain item (x in y)
NOT IN - return True if the queried list doesn't present in the item (x not in y)
```

````python
x = 3
y = 4
z = [1,2,3,4,5,6]

print(x is y) #False
print(x is not y) #True
print(x in z) #True
print(y not in z) #False - True if the value is not present in the list.
````

```py
#Logical Operator
- These operator are used in combining the condition.

AND - Both conditions should be True
OR - Either of the condition should be True to produce result True.
NOT - reverse the result. If True -> False;False -> True.
```

````python
print(3>2 and 5>4) #True - as both are true
print(4<8 and 9>8) #False - One or both are False
print(5==5 or 5>6) #True - as one of the condition is true.
print(6!=6) #False - as it will reverse the output.
````
