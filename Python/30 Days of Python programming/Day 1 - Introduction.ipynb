```
Interactive vs Script mode:

Interactive:
- execution of the script can be done line by line.
- code cannot be saved
- quick output
- debug instantly

cons:
- not suitable for huge programs

Best Usage:
- when you need to try new instant code/checks


Script:
- programs are written with .py extensions and executed using the file in the terminal/run button.
- Execution of the code is done in single go
- code can be saved.
- good debug feature
- GITHUB and Version control integration
- Code reusable

cons:
- not suitable for quick output checks.

Best Usage:
- Good for building projects / developement / applications

Ex: first_code.py
       |
    terminal: python first_code.py
       |
    hello world

```
