# Datatypes
- It defines what a variable is holding
- Int, str, bool, complex, float

````python
print(type(5)) #Integer
print(type(2.5)) #Float
print(type('hello')) #string
print(type(2+3j)) #complex
print(type(True)) #Boolean
print(type([1,2,3])) #list
print(type((1,2,3,4))) #tuple
print(type({1:'hi',2:'bye'})) #Dictionry
print(type({1,2,3,4})) #set
print(type(zip([1,2,3],[4,5,6]))) #Zip
````

Casting: The process of converting one data type to another data type using built in function like str(),int(), float(), this will raise an error is we try to concatenate two variables of different data types.

- We cannot add one integer and one string having numbered value. so to solve this we need to convert the string data type to integer data type.

````python
#Type casting from str to int
a = 10
b = '20'
c = a + int(b)
print(c)
````
