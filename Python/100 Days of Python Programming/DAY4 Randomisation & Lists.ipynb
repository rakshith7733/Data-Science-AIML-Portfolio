```python
Data structure -- the way of storing and organizing the data.

Lists: They are dynamic array that can store elements of different types. 
       most commonly used built-in data structure as it is flexibility and easy to use.
       It is a Built-in data structure in python which are mutable or changable, ordered sequence of elements of collection.
defined by [] or list() constructor.

ex: list = [1,2,'rak','shith',8]
    
    #using list constructor.
    constructed_list = list((1,2,3,4))
----------------------------------------------------------------------------------------------------------------------------------------
Scenario:
Until now we have seen how to store data individually. But, for storing huge or more amount of data in organized way we can use list. 
This because they are mutable, sequence format.

Pros:
    Best for Dynamic arrays & ordered collections.
    Mutable -- elements can be changed.
    supports indexing and slicing of list.
Use:
    when frequent appends and read need.
    when you store different data types together.

Cons:
    Not ideal for frequent insertions / deletions in the middle. (Use Deque data structures.)
    Not efficency for quick lookups. (Use set or dict)

Common list methods:

                                                               Adding elements~
1. Append(value) -- adds element at the end.
         Ex: List = [1,2,3]
            List.append(4)     #O/p --> [1,2,3,4]

2. Insert(index, value) -- adds element at mentioned index.
         Ex: List = [1,2,3]
             List.Insert(2,4) #O/P -> [1,2,4,3]

3. Extend(iterable) -- extends the existing list.
         Ex: List = [1,2,3]
             List.extend([4,5,6])  #O/P --> [1,2,3,4,5,6]

                                                              Removing elements~
4. Pop(index) -- removes and return the eleement.
         Ex: List = [1,2,3,4]
             List.pop()    #removes last index value. --> [1,2,3]
             List.pop(2)   #removes index 2 value. --> [1,2,4]

5. remove(value) -- remove the first occurance value from a string.
         Ex: list = [1,2,3]
             list.remove(3)  #O/p --> [1,2]

6. Clear() -- removes all elements from the list.
         Ex: list = [1,2,3]
             list.clear()  #O/P --> []

                                                              Searching elements~
7. index(value) -- return the index of first occurance of the list.
         Ex: list = [1,2,3,2]
             list.index(2)  #O/p --> 3

8. count(value) -- count occurance of list.
         Ex: list = [1,2,3,2,5,6,2,7,9,2]
             list.count(2)     #O/p --> 4

9. In keyword -- check is that present in the list.
         Ex: list = [1,2,3,4]
             print(3 in list)    #True.

                                                              Sorting Elements~
10. sort() -- sorts the list in place.
         Ex: list = [9,7,5,2]
             list.sort()      #O/p --> [2,5,7,9]

11. sorted(iterable) -- returns a new sorted list.
         Ex: list = [9,5,1,8,4]
             list1 = sorted(list)   #O/p -->[1,4,5,8,9]

12. reversed(iterable) -- returns an iterator for the reversed list.
         Ex: list = [3,5,6,8,9,4,2,7]
             list.reverse()          #O/p --> [7, 2, 4, 9, 8, 6, 5, 3]

13. copy() -- create a shallow copy
         Ex: list = [1,2,3,5,6]
             list1 = list.copy()     #O/p --> list1 = [1,2,3,5,6]

14. [:] -- slicing the list and also creates a copy.
```
