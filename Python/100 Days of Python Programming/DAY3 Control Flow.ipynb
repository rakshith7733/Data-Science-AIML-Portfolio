```python
Conditional statements: 

These are used to execute certain block of code based on the specific conditions. These statements help control the flow of the program, make it behave differently in different situations.

IF-Else statement:

    If Condition:
       do this
    else:
       do this
```

````python
#conditional statement.

water_level = 80
if water_level < 80:
    print("Continue")
else:
    print("Drain water")
````
