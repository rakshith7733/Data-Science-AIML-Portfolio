````python
#Rollercoster ride
#Condition height >= 120 & age less than 12, 12-18 and over 18. add need picture of ride then add $3. Aged ticket(age = 45-55) will cost 0 for ride.
print("Welcome to the rollercoaster!")
height = int(input("Enter your height? in CM?"))

bill = 0
ticket = ""

if height >= 120:
    print("You can Ride.")

    #Age
    age = int(input("Enter your Age: "))
    if age >= 45 and age <= 55:
        bill = 0
        ticket = 'Elder Ticket'
        print("You need to pay have a free ride.")
    elif age <= 12:
        bill += 5
        ticket = 'Child Ticket'
    elif age <= 18:
        bill += 7
        ticket = 'Younger Ticket'
    elif age > 18 and age < 45:
        bill += 12
        ticket = 'Adult Ticket'

    photo = input("Do you need a Picture. Y or N: ").lower()
    if photo == 'y':
        bill += 3
    
else:
    print("You can't ride...")

print(f"You opt for {ticket} and cost of your ride is ${bill}. ")
````

```python
~ Pizza Delivery.

Pizza scenario requirements:
1. small pizza = 15, medium pizza = 20, large pizza = 25
2. pepperoni for small = 2, pepperoni for medium & large = 3
3. extra cheese = 1 for any pizza
```

````python
print("Welcome to Python Pizza Deliveries!")

size = input("What size pizza do you want? S,M, or L: ").lower()

#If user try to choose correct input or not.
if size == 's' or size == 'm' or size == 'l':
    
    pepperoni = input("Do you want pepperoni on your pizza? Y or N: ").lower()

    #Creating variable to store respective elements.
    bill = 0
    pizza_size = ""
    pepperoni_select = ""
    cheese_select = ""

    #Size option
    if size == 's':
        bill += 15
        pizza_size = "Small"
        if pepperoni == 'y':
            bill += 2
            pepperoni_select = "Yes"
        else:
            pepperoni_select = "No"

    elif size == 'm':
        bill += 20
        pizza_size = "Medium"
        if pepperoni == 'y':
            bill += 3
            pepperoni_select = "Yes"
        else:
            pepperoni_select = "No"

    elif size == 'l':
        bill += 25
        pizza_size = "Large"
        if pepperoni == 'y':
            bill += 3
            pepperoni_select = "Yes"
        else:
            pepperoni_select = "No"

    #Cheese option
    extra_cheese = input("Do you want extra cheese? Y or N: ").lower()
    if extra_cheese == 'y':
        bill += 1
        cheese_select = "Yes"
    else:
        cheese_select = "No"


    print(f"You have choose '{pizza_size}' Pizza and you selected '{pepperoni_select}' for Pepperoni & '{cheese_select}' Cheese. \n Your Final Bill: ${bill}.")

else:
    print("You have choose wrong input. Refreshing inputs.")
````
