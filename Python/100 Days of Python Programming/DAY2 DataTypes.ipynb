```python
What is Data Type?
It defines a varibale what data it will be holding and operation that can be performed.

Premitive Data types: 
                    They represent simple values and are not composed of other data types.
                    Stores single values, 
                    fast and efficient.


1. Intgers(Int) - Whole numbers. 
                  Ex:2 400 50

2. Float(float) - numbers with decimal points.
                  Ex: 3.5  456.6

3. Boolens(Bool) - logical value indicating True / False.

Non-Premitive Data types:
                         These are complex data structures that can store multiple values or behaviours.
                         Stores multiple values,
                         can have methods and functions.
                         
4. Strings(Str) - ordered sequence of characters.
                  Ex: "hello" "good"

5. Lists(list) - ordered sequence of objects.
                  Ex: [1,"hello", 4,"good"]

6. Dictinaries(Dict) - unordered key:value pairs.
                  Ex: {"A":"Excellent"
                       "B":"Good"
                       "C":"Average"
                       }

7. Tuples(Tup) - ordered immutable sequence of objects.
                Ex: (10,"hello",4,"good")

8. Sets(set) - unordered collection of unique objects.
                Ex: ("A","B")

Subscripting: The ability to pull a particular elements from the string using the indexing is know as subscripting method.
Ex: Print('Hello'[1])    #here the index method is used to pull the character from string.
    O/P: e
```

```python
They are two types of Data types:
1. Premitive - single valued types

Int     -> Whole numbers --> 10
Float   -> decimal numbers --> 3.5
complex -> Complex numbers --> 2 + 3j
bool    -> True / False --> True
str     -> text values --> "Hello"

2. Non- Premitive -- multiple valued

List       (Mutable, Ordered)         -> lst=[1,2,"go",5]
tuple      (unmutable, Ordered)       -> tup=(1,"go",3,"good")
set        (Unordered, unique values) -> set=(1,2,3)
Dictinoary (key: value pair)          -> dict={"A":"excellent", "B":"Good", "C": "Average"}

*Mutable - Can change after creation in place (list, set, dict)
*Immutable - cannot change after creation (int,float, bool, str, tuple)
```

````python
#Subscritping
print('Hello'[1])

#binary
new = b'good'
print(new)

#string
print("123" + "345")

#integers - whole numbers
print(123 + 345)

#large integers
print(123_456_768)  #here underscore refers as , we use to large digits.

#float point number
print(3.14)

#boolen numbers
print(True)
````

           An exception that occurs when the data types of an object in an operation is inappropriate. this occurs when the operations performed on the object is incorrect or not supported for the object.

```python
Round() function: 
                 when we want to round the decimal number to nearest whole number using traditional mathematical way. where anything over .5 rounds up and anything below .5 round down.
round(value,no of digits)
```

````python
#BMI Calculator
height = 1.65
weight = 84

bmi = round(weight / (height * height),2)
print(f'Your BMI Score is {bmi}')
````

````python
#DAY2 Project: Tip Calculator
print("Welcome to the Tip Calculator!")

bill = float(input("What was the total bill? "))  #124.54
tip = int(input("How much tip would you like to give? 10,12, or 15? ")) #12
people = int(input("How many people to split the bill? ")) #5

#logic
tip_percent = tip /100  #1.12
bill_af_tip = bill + (bill * tip_percent)
split = bill_af_tip / people
final_bill = round(split,2)

#Display of bill after splitting.
print(f"Each person should pay: ${final_bill}") #27.90
````
