```python

- This is why python uses more memory than any other programming everything is stored in objects.

and later we assign other value to X variable. it will chose the second value.

````python
#Dynamic typing in Python:
#same variable but can be changed according to requirement and it will adapt to data type that value is stored.

x = 5        #Int
print(type(x))
print(id(x))
x = "Helo"   #Str
print(type(x))

#To check memory address
print(id(x))
````

```python
Errors Types:
1.  Syntax Error: 
The set of rules that are defines how a python program will be written and interpreted.

Ex: Missing, Mispelled, keywords, Incorrect Indentation, missing parantheses.

2. Indentation Error:
Python expects proper spacing(Indentation), especially in blocks like IF, FOR, DEF.

Ex: IF True:
    print(x) #missing Indentation

3. Name Error:
A variable or function that has not be defined.

Ex: num = 5
    print(x)  #printing variable which is not defined.

4. Runtime Error:
Code runs but crashes during execution.

Ex: x = 10/0
    print(x) #code will run but produce infinity loop which crashes the terminal.

5. Type Error: 
These errors occur when you are using the wrong data types, invalid type annotations.

Ex: len(1234) #we are checking the length of a number, which is not applicable.

6. Index Error:
You are accessing an index that do not exists.

Ex: arr = [1,2,3]
    print(arr[5])  #out of range.

7. Value error:
           when an object is given the incorrect value, the interpretor will raise ValueError. This occurs when the value is incorrect for the intended operations or if the value is missing altogether.
           
Ex: Print(int("abc") + int("456"))

```
